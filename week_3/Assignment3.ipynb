In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

Steps:

To calculate the errors in the mean:

$\sigma_{\overline{x}} = \sigma_x  /\sqrt{N}$

And the errorbar for the 95% confidence interval is 1.95 times this standard deviation

Calculate whether y-value is below, within, or above the 1-sigma range from the mean. Implement the following colorbar scheme:

* red if data is definitely above the y-value, 
* blue if data is definitely below the y-value,
* white if y-value is within the 1-sigma range.

```python
def easiest(y_value, df):
    n_samples = len(df.columns.values)
    df_indices = df.index.values
    df_mean = df.mean(axis=1)
    df_err = 1.96*df.std(axis=1)/np.sqrt(n_samples)
    
    years_above_mask = (df_mean - df_std) > y_value
    years_below_mask = (df_mean + df_std) < y_value

    df['color'] = 'w'
    df.loc[years_above_mask,'color'] = 'r'
    df.loc[years_below_mask,'color'] = 'b'
    
    fig = plt.figure(figsize=(4,5))
    ax  = plt.gca()
    bars = ax.bar(df_indices, df_mean,  
                  yerr = df_err, ecolor='k',
                  align='center', width=.75, color=df['color'], edgecolor='k')
    ax.axhline(y_value,c='k')

    xlabels = df.index.values
    plt.xticks(xlabels, xlabels)
    plt.ylabel('Mean value')

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    ax.tick_params(top = 'off', right = 'off', left='on')
    plt.tight_layout()
    plt.show()
```

# Harder option
Implement the bar coloring as described in the paper, where the color of the bar is actually based on the amount of data covered (e.g. a gradient ranging from dark blue for the distribution being certainly below this y-axis, to white if the value is certainly contained, to dark red if the value is certainly not contained as the distribution is above the axis).

I want to calculate the z-value for each bar, then calculate the probability from the normal distribution.

Question: How do I create a colormap based on a handful of values? Want the colorbar to go from lowest to highest. Maybe label it with qualitative things like "likely below, likely above, likely in range". 

Create a colorbar on the right that goes from 0 to 1. Use the `mpl_il` from Week 3 notes.

```python
def create_rgb(z):
    # Transform z to probability
    prob = st.norm.cdf(z)
    rgb = cm.get_cmap(plt.get_cmap('RdBu_r'))(prob)
    
    return rgb
```

```python
def harder(y_value, df, f, ax1, ax2):    
    n_samples = len(df.columns.values)
    df_indices = df.index.values
    df_mean = df.mean(axis=1)
    df_err = 1.96*df.std(axis=1)/np.sqrt(n_samples)
    
    z_value = (df_mean-y_value)/df_err
    colors = create_rgb(z_value)    
    
    norm = mpl.colors.Normalize(vmin=0,vmax=1)
    mpl.colorbar.ColorbarBase(ax2, cmap=cm.RdBu_r,
                              norm=norm, orientation='horizontal')
    
    bars = ax1.bar(df_indices, df_mean, 
                  yerr = df_err, ecolor='k',
                  align='center', width=.75, color=colors, edgecolor='k')
    ax1.axhline(y_value,c='k')
    ax1.set_ylim([0, 65000])

    xlabels = df.index.values
    ax1.set_xticks(xlabels)
    ax1.set_xticklabels(xlabels)
    ax1.set_ylabel('Mean value')

    ax1.spines['right'].set_visible(False)
    ax1.spines['top'].set_visible(False)

    ax1.tick_params(top = 'off', right = 'off', left='on')
    ax2.tick_params(top = 'off', right = 'off', left = 'off', bottom = 'off', labelbottom='off')

    ax2.annotate('below', xy=(0.2,0.15), xycoords='figure fraction',fontsize=10)
    ax2.annotate('contained', xy=(0.525,0.15), xycoords='figure fraction',fontsize=10)
    ax2.annotate('above', xy=(0.87,0.15), xycoords='figure fraction',fontsize=10)
    ax1.annotate('y_value={:.0f}'.format(y_value), xy=(0.45,0.9), xycoords='figure fraction',fontsize=10)
    
    plt.tight_layout()
    return
```

```python
fig = plt.figure(figsize=(4,3),facecolor='w')
gs = gridspec.GridSpec(2,1,height_ratios=[10,1])
ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])

harder(40000,df, fig, ax1, ax2);
plt.savefig('bar_chart.png',facecolor='w',edgecolor='w')
```
